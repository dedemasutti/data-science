## Importando as bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Carregando os dados a serem analisados

```python
# Carregar o arquivo de dados (por exemplo, um arquivo CSV)
dados = pd.read_csv('/content/dataset.csv', sep = ';')

#Caso seja um arquivo .json, por exemplo, o método será:
#data = pd.read_json
```

## Explorando os dados

```python
# Visualizar as cinco primeiras linhas do dataframe
dados.head()
```

```python
# Visualiza as cinco últimas linhas do dataframe
dados.tail()
```

```python
# Obter informações gerais sobre o dataframe
dados.info()
```

```python
# Verificar estatísticas descritivas dos dados numéricos
dados.describe()
```

```python
# Verificar a contagem de valores únicos em cada coluna
dados.nunique()
```

```python

# Verificar a presença de valores ausentes
dados.isnull().sum()
```

```python
# Podemos ver os tipos de dados de um dataframe
dados.dtypes
```

```python
# Calcular a matriz de correlação entre as colunas numéricas
correlation_matrix = dados.corr()
```

## Análise exploratoria com graficos

```python
# Plotar um mapa de calor da matriz de correlação
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn')
plt.title('Matriz de Correlação')
plt.show()

# Analisar a distribuição de uma variável numérica
plt.figure(figsize=(8, 6))
sns.histplot(dados['Valor'], kde=True)
plt.title('Distribuição da Coluna Numérica')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()

# Analisar a relação entre duas variáveis numéricas
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Valor', y='Area', data=dados)
plt.title(' ações do ITAU e ações da JHFS')
plt.show()

# Analisar a distribuição de uma variável categórica
plt.figure(figsize=(8, 6))
sns.countplot(dados['Area'])
plt.title('Distribuição da Coluna Categórica')
plt.xlabel('Categoria')
plt.ylabel('Contagem')
plt.show()

# Analisar a relação entre uma variável categórica e uma variável numérica
plt.figure(figsize=(10, 8))
sns.boxplot(x='Valor', y='Dist_Praia', data=dados)
plt.title('Relação entre Coluna Categórica e Coluna Numérica')
plt.xlabel('Categoria')
plt.ylabel('Valor Numérico')
plt.show()

# Analisar a distribuição conjunta de várias variáveis numéricas
sns.pairplot(dados, vars=['Valor', 'Area', 'Dist_Praia'])
plt.show()
```

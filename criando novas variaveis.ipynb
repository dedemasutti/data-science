# Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.shape[1]
```

```python
dados.head(10)
```

```python
# Criamos uma nova variável com o nome de  Valor Bruto. Essa nova variavel é a soma das variaveis: Valor, Condominio e IPTU.
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
#Em seguida, criaremos outra variável que armazenará o valor por metro quadrado, chamaremos essa nova variável de Valor m2. Iremos coletar Valor e dividi-lo por Area.
dados['Valor m2'] = dados['Valor']/dados['Area']
```

```python
dados.head(10)
```

```python
#Vamos arredondar os valores da variavel Valor m2. Fazemos isso com o método round(). Esse método arredonda os valores float.
dados['Valor m2'] = dados['Valor m2'].round(2)
dados.head(10)
```

```python
# Próximo passo é criar a váriavel Valor Bruto m2.
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
dados.head(10)
```

```python
#Criaremos uma lista com a identificação das casas para usarmos na próxima nova variavel.
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
#criaremos efetivamente a nova variável que chamaremos de Tipo Agregado. Depois, evocaremos dados e passaremos a variável Tipo, seguido do método apply(). Esse método pertmite que apliquemos uma função à cada registro do DataFrame. 
#Para tanto, criaremos uma função lambda
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(10)
```

# Excuindo variaveis

```python
# Criamos um banco de dados para darmos o primeiro exemplo
dados_auxiliares = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
dados_auxiliares.head(10)
```

```python
# Usaremos a palavra  del para excluir a váriavel Valor Bruto.
del dados_auxiliares['Valor Bruto']
```

```python
dados_auxiliares
```

```python
# Método pop é outro jeito de excluirmos váriaveis

dados_auxiliares.pop('Valor Bruto m2')
```

```python
dados_auxiliares
```

```python
#Uma maneira de excluirmos múltiplas variáveis é utilizando o método drop().
# método drop(), precisaremos especificar o eixo em que se dará a exclusão 
#uma vez que ele pode atuar especificamente em linhas ou colunas. Então, escreveremos axis = 1, afinal o eixo 0 corresponde às linhas e o 1 colunas. Vamos usar inplace como True.

dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep = ';', index = False)
```
